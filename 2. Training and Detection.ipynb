# 8. Load Train Model From Checkpoint

```python
import os
import tensorflow as tf
from object_detection.utils import label_map_util
from object_detection.utils import visualization_utils as viz_utils
from object_detection.builders import model_builder
from object_detection.utils import config_util
```

```python
# Load pipeline config and build a detection model
configs = config_util.get_configs_from_pipeline_file(files['PIPELINE_CONFIG'])
detection_model = model_builder.build(model_config=configs['model'], is_training=False)

# Restore checkpoint
ckpt = tf.compat.v2.train.Checkpoint(model=detection_model)
ckpt.restore(os.path.join(paths['CHECKPOINT_PATH'], 'ckpt-12')).expect_partial()

@tf.function
def detect_fn(image):
    image, shapes = detection_model.preprocess(image)
    prediction_dict = detection_model.predict(image, shapes)
    detections = detection_model.postprocess(prediction_dict, shapes)
    return detections
```

# 9. Real Time Detections from your Webcam

```python
import cv2
import numpy as np
import tensorflow as tf
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
import pygame
from pygame.locals import *
```

```python
pygame.init()
```

```python
def open_canvas(color):
    pygame.init()
    screen = pygame.display.set_mode((1250, 750))
    pygame.display.set_caption("Virtual Pen")

    canvas = np.zeros((2,2, 3), dtype=np.uint8)

    clock = pygame.time.Clock()
    drawing = False

    while True:
        clock.tick(60)
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                return
            elif event.type == pygame.MOUSEBUTTONDOWN:
                drawing = True
            elif event.type == pygame.MOUSEBUTTONUP:
                drawing = False
            elif event.type == pygame.MOUSEMOTION:
                if drawing:
                    if color == 'red':
                        pygame.draw.circle(screen, (255, 0, 0), event.pos, 5)
                        pygame.draw.circle(pygame.surfarray.make_surface(canvas), (255, 0, 0), event.pos, 5)
                    elif color == 'blue':
                        pygame.draw.circle(screen, (0, 0, 255), event.pos, 5)
                        pygame.draw.circle(pygame.surfarray.make_surface(canvas), (0, 0, 255), event.pos, 5)
        screen.blit(pygame.surfarray.make_surface(canvas), (0, 0))
        pygame.display.update()
```

```python
lower_red = np.array([161,155,84]) 
upper_red = np.array([179, 255, 255]) 

lower_blue =np.array([94,80,2])
upper_blue =np.array([126,255,255])


cap = cv2.VideoCapture(0)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

while cap.isOpened(): 
    
    
    ret, frame = cap.read()
    image_np = np.array(frame)
    
    input_tensor = tf.convert_to_tensor(np.expand_dims(image_np, 0), dtype=tf.float32)
    detections = detect_fn(input_tensor)
    
    
    success, video = cap.read() # Reading webcam footage
    img = cv2.cvtColor(video, cv2.COLOR_BGR2HSV) # Converting BGR image to HSV format

    mask_red = cv2.inRange(img, lower_red, upper_red) 
    mask_blue = cv2.inRange(img, lower_blue, upper_blue)

    mask_contours_red, hierarchy_red = cv2.findContours(mask_red, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE) 
    mask_contours_blue, hierarchy_blue = cv2.findContours(mask_blue, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE) 

    detected_object = 0

    # Finding position of all red contours
    if len(mask_contours_red) != 0:
        for mask_contour_red in mask_contours_red:
            if cv2.contourArea(mask_contour_red) > 500:
                x, y, w, h = cv2.boundingRect(mask_contour_red)
                cv2.rectangle(video, (x, y), (x + w, y + h), (0, 0, 255), 3) #drawing rectangle
                detected_object = 'red'

    # Finding position of all blue contours
    if len(mask_contours_blue) != 0:
        for mask_contour_blue in mask_contours_blue:
            if cv2.contourArea(mask_contour_blue) > 500:
                x, y, w, h = cv2.boundingRect(mask_contour_blue)
                cv2.rectangle(video, (x, y), (x + w, y + h), (255, 0, 0), 3) #drawing rectangle
                detected_object = 'blue'
                
    cv2.imshow("window", video) # Displaying webcam image
    
    num_detections = int(detections.pop('num_detections'))
    detections = {key: value[0, :num_detections].numpy()
                  for key, value in detections.items()}
    detections['num_detections'] = num_detections

    # detection_classes should be ints.
    detections['detection_classes'] = detections['detection_classes'].astype(np.int64)

    label_id_offset = 1
    image_np_with_detections = image_np.copy()

    viz_utils.visualize_boxes_and_labels_on_image_array(
                image_np_with_detections,
                detections['detection_boxes'],
                detections['detection_classes']+label_id_offset,
                detections['detection_scores'],
                category_index,
                use_normalized_coordinates=True,
                max_boxes_to_draw=5,
                min_score_thresh=.8,
                agnostic_mode=False)

    cv2.imshow('object detection',  cv2.resize(image_np_with_detections, (800, 600)))
    
    if len(detections['detection_scores']) > 0 and detections['detection_scores'][0] > 0.8:
        open_canvas(detected_object)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        cap.release()
        cv2.destroyAllWindows()
        break
```
